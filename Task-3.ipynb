## Data Visualization Using Python

### Create a bar chart and a line chart using Matplotlib to visualize data from a Pandas DataFrame. Customize the charts with labels, titles, and legends.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/DELL/Downloads/Datasets/householdtask3.csv")
```

```python
# Scatter plot with year against own 
plt.scatter(data['year'], data["own"])

# Adding title to the plot 
plt.title("Scatter Plot")

# Setting the x and y Label
plt.xlabel('year')
plt.ylabel('own')

# showing the result
plt.show()
```

```python
# Line Chart with age against income
plt.plot(data['age']) 
plt.plot(data['income'])

# Adding title to the plot 
plt.title("Line Chart")

# Setting the x and y Label 
plt.xlabel('age') 
plt.ylabel('income')

# showing the result 
plt.show()
```

```python
# Bar plot 
plt.bar(data['year'], data['own_prop'])

# Adding title to the plot 
plt.title("Bar")

# Setting the x and y Label 
plt.xlabel('year') 
plt.ylabel('own_prop')

# showing the result 
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

# Adding title to the plot
plt.title("Histogram")

# Display Plot
plt.show()
```
